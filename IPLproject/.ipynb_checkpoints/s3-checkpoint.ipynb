```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

```

```python
train.describe()
```

```python
train.info()
```

```python
train = train.drop(['prod_desc', 'prod_long_desc', 'set_name', 'prod_id'], axis='columns')
train = train.iloc[:,1:]

test = test.drop(['prod_desc', 'prod_long_desc', 'set_name', 'prod_id'], axis='columns')
test = test.iloc[:,1:]

train.head()
```

```python
train.size
```

```python
train.shape
```

```python
theme_stats = train['theme_name'].value_counts(ascending = False)
```

```python
theme_less_than_50 = theme_stats[theme_stats<=50]
```

```python
train.theme_name = train.theme_name.apply(lambda x: 'other' if x in theme_less_than_50 else x)
test.theme_name = test.theme_name.apply(lambda x: 'other' if x in theme_less_than_50 else x)

test['theme_name'].value_counts(ascending = False)
```

```python
import matplotlib as plt
train.hist(bins=50, figsize=(20,15))
```

```python
grouped = train.groupby('theme_name')['list_price'].mean()
grouped.head()
```

```python
#Create Dictionary
theme_price_dict = dict(grouped)
#Create new column with dictionary
train['Theme_Avg_Price'] = train.theme_name.map(theme_price_dict)
test['Theme_Avg_Price'] = test.theme_name.map(theme_price_dict)
train = train.dropna()

test.head()
```

```python
def avg_feat(cfeat, nfeat):
    new_col = '{}_Avg_{}'.format(cfeat, nfeat)
    grouped = train.groupby(cfeat)[nfeat].mean()
    train[new_col] = train[cfeat].map(dict(grouped))
    test[new_col] = test[cfeat].map(dict(grouped))
    
cat_feats = ['ages', 'review_difficulty', 'country']
num_feats = ['list_price', 'num_reviews', 'piece_count', 'play_star_rating', 'star_rating', 'val_star_rating']
for cfeat in cat_feats:
    for nfeat in num_feats:
        avg_feat(cfeat, nfeat)
        
train.head(2)
```

```python
train['Has_Trademark'] = train.theme_name.str.contains('™')
train.Has_Trademark.value_counts(normalize=True)

test['Has_Trademark'] = train.theme_name.str.contains('™')
test.Has_Trademark.value_counts(normalize=True)
```

```python

test["num_reviews"].fillna(train["num_reviews"].median(), inplace=True)
test["play_star_rating"].fillna(train["num_reviews"].mean(), inplace=True)
test["star_rating"].fillna(train["star_rating"].mean(), inplace=True)
test["val_star_rating"].fillna(train["val_star_rating"].mean(), inplace=True)
test["review_difficulty_Avg_list_price"].fillna(train["review_difficulty_Avg_list_price"].mean(), inplace=True)
test["review_difficulty_Avg_num_reviews"].fillna(train["review_difficulty_Avg_num_reviews"].mean(), inplace=True)
test["review_difficulty_Avg_val_star_rating"].fillna(train["review_difficulty_Avg_val_star_rating"].mean(), inplace=True)
test["review_difficulty_Avg_piece_count"].fillna(train["review_difficulty_Avg_piece_count"].mean(), inplace=True)
test["review_difficulty_Avg_play_star_rating"].fillna(train["review_difficulty_Avg_play_star_rating"].mean(), inplace=True)
test["review_difficulty_Avg_star_rating"].fillna(train["review_difficulty_Avg_star_rating"].mean(), inplace=True)
test["review_difficulty_Avg_val_star_rating"].fillna(train["review_difficulty_Avg_val_star_rating"].mean(), inplace=True)
test.describe()
```

```python
"4-99".split('-')[1]
```

```python
def strip_age(age):
    if '+' in age :
        min = age.strip('+')[0]
        max = 99
    if '-' in age :
        splitted = age.split('-')
        min = splitted[0]
        max = splitted[1]
        if '½' in min :
            min = float(min.split('½')[0]) +0.5
            
    return pd.Series([float(min), float(max)])

print(len(["min_age", "max_age"]), len(strip_age('18+')))

train[["age_min", "age_max"]] = train.ages.apply(strip_age)
test[["age_min", "age_max"]] = test.ages.apply(strip_age)
```

```python
train = train.drop(['ages'], axis='columns')
test = test.drop(['ages'], axis='columns')
```

```python
test.head()
```

```python
from sklearn.preprocessing import LabelBinarizer, LabelEncoder

encoder = LabelEncoder()
binarizer = LabelBinarizer()

train["review_difficulty"] = encoder.fit_transform(train['review_difficulty'])
train["Has_Trademark"] = encoder.fit_transform(train['Has_Trademark'])
test["review_difficulty"] = encoder.fit_transform(test['review_difficulty'])
test["Has_Trademark"] = encoder.fit_transform(test['Has_Trademark'])
```

```python
test
```

```python
X = train.drop(['theme_name', 'country', 'list_price'], axis ='columns')
topredict = test.drop(['theme_name', 'country'], axis ='columns')
y = train.list_price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)
```

```python
dt = DecisionTreeRegressor(criterion='mse')
dt.fit(X_train,y_train)
dt.score(X_test,y_test)
```

```python
predict = dt.predict(topredict)
np.savetxt("predict.csv", predict, fmt="%d", delimiter="\n")
```
